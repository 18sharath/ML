```python
X_train,X_test,y_train,y_test=X[:60000],X[60000:],y[:60000],y[60000:]
shuffling_index=np.random.permutation(6000)
shuffling_index
```

```python
# shuffeling the traiing data set to make 
#  Some learning algorithms are sensitive to the order 
X_train,y_train=X_train[shuffling_index],y_train[shuffling_index]
```

```python
X_train
```

```python
#  target vectors for classification task to distinguish whether its 5 or not
y_train_5=y_train==5
y_test_5=y_test==5
```

```python
y_train_5
```

```python
from sklearn.linear_model import SGDClassifier
sgd_clf=SGDClassifier()
sgd_clf.fit(X_train,y_train_5)
y_train_pred=sgd_clf.predict(X_test)
```

```python
print(sgd_clf.predict([X[0]]))
```

```python
# performace measure
# measuring accuracy using cross-validation
# Divide the dataset into k subsets (called folds). 
# train the model on k-1 times on model and remianing folds on test
# The average of the scores from each iteration is the cross-validation score.
from sklearn.model_selection import cross_val_score
cross_val_score(sgd_clf,X_train,y_train_5,cv=4,scoring='accuracy')
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test_5, y_train_pred)
```

```python
from sklearn.metrics import precision_score, recall_score
precision_score(y_test_5, y_train_pred)
```

```python
# precision and recall trade-off
y_score=sgd_clf.decision_function([X[11]])
y_score
```

```python
threshold=0
y_some_digit_pred=(y_score>threshold)
y_some_digit_pred
```
